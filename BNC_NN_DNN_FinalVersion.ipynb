```python
import pandas as pd
path = "/content/drive/MyDrive/ml_dataset/clean_dataset.csv"
df = pd.DataFrame

data_set = pd.read_csv(path)
```

```python
temp_df1 = data_set[data_set['category'] == 'bangladesh'][:5000]
temp_df2 = data_set[data_set['category'] == 'international'][:5000]
temp_df3 = data_set[data_set['category'] == 'entertainment'][:5000]
temp_df4 = data_set[data_set['category'] == 'economy'][:5000]
temp_df5 = data_set[data_set['category'] == 'opinion'][:5000]
temp_df6 = data_set[data_set['category'] == 'technology'][:5000]
temp_df7 = data_set[data_set['category'] == 'life-style'][:5000]
temp_df8 = data_set[data_set['category'] == 'sports'][:5000]

temp_data = pd.concat([temp_df1,temp_df2,temp_df3,temp_df4,temp_df5,temp_df6,temp_df7,temp_df8])
```

```python
temp_data.head(5)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(15, 10))
count_classes = temp_data.category.value_counts()
count_classes.plot(kind = 'bar', rot=0)

plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
X = temp_data.loc[:,("content")]
Y = temp_data.loc[:,("category")]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2)
```

```python
x_train[:10]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(lower=True)
tokenizer.fit_on_texts(x_train.values.astype(str))
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
max_length = 500
trunc_type='post'
padding_type='post'

x_train_sequences = tokenizer.texts_to_sequences(x_train.values.astype(str))
x_train_padded = pad_sequences(x_train_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

x_test_sequences = tokenizer.texts_to_sequences(x_test.values.astype(str))
x_test_padded = pad_sequences(x_test_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
print(x_train[0:1])
print(x_train_padded[0])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

en_y_train = le.fit_transform(y_train)
en_y_test = le.transform(y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
en_y_train = en_y_train.reshape((en_y_train.shape[0], 1))
y_train_ohe = encoder.fit_transform(en_y_train)
en_y_test = en_y_test.reshape((en_y_test.shape[0], 1))
y_test_ohe = encoder.fit_transform(en_y_test)
```

```python
print(x_train_padded.shape)
# print(en_y_train.shape)
print(y_train_ohe.shape)
print(x_test_padded.shape)
print(y_test_ohe.shape)
```

```python
from tensorflow.keras.layers import Dense, Dropout, Embedding, AveragePooling1D,Flatten, MaxPooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=8, input_length=max_length))
model.add(AveragePooling1D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(8, activation='softmax'))  # add output layer with 8 units and 'softmax' activation

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
history = model.fit(x_train_padded, y_train_ohe, epochs=10, batch_size= 512, validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience =3)])
```

```python
model.save('Dense_NN')
```

```python
accr = model.evaluate(x_test_padded,y_test_ohe)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```

```python
import tensorflow as tf

model_hybrid = tf.keras.Sequential()

model_hybrid.add(Embedding(input_dim=vocab_size, output_dim=8, input_length=max_length))

model_hybrid.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)))

model_hybrid.add(tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64, return_sequences=True)))

model_hybrid.add(Flatten())

model_hybrid.add(tf.keras.layers.Dense(128, activation='relu'))
model_hybrid.add(tf.keras.layers.Dense(8, activation='softmax'))

```

```python
model_hybrid.summary()
```

```python
model_hybrid.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history_hybrid = model_hybrid.fit(x_train_padded, y_train_ohe, epochs=10, batch_size= 512 , validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience =2, restore_best_weights=True)])
```

```python
model.save('Hybrid_RNN.hi5')
```

```python
import matplotlib.pyplot as plt
plt.title('Accuracy Vs Validation Accuracy')
plt.plot(history_hybrid.history['accuracy'], label='accuracy')
plt.plot(history_hybrid.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show();
```

```python
accr2 = model_hybrid.evaluate(x_test_padded,y_test_ohe)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr2[0],accr2[1]))
```
